### 1.3 Find the top 10 cuisines served by maximum number of restaurants in Delhi NCR and rest of India.

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import pandas as pd
pd.options.mode.chained_assignment = None

def func(i):
    if i == 'New Delhi' or i == 'Ghaziabad' or i == 'Noida' or i == 'Gurgaon' or i == 'Faridabad':
        return 'Delhi-NCR'
    else:
        return i

    
data = pd.read_csv("zomato.csv", encoding = "ISO-8859-1")

rest_of_india = data[data['Country Code'] == 1]
rest_of_india = rest_of_india[rest_of_india['City'] != 'New Delhi']
rest_of_india = rest_of_india[rest_of_india['City'] != 'Ghaziabad']
rest_of_india = rest_of_india[rest_of_india['City'] != 'Noida']
rest_of_india = rest_of_india[rest_of_india['City'] != 'Gurgaon']
rest_of_india = rest_of_india[rest_of_india['City'] != 'Faridabad']

delhi_ncr = data[data['Country Code'] == 1]
delhi_ncr['City'] = delhi_ncr.City.apply(func)
delhi_ncr = delhi_ncr[delhi_ncr['City'] == 'Delhi-NCR']

delhi_ncr_cuisine = dict()
rest_of_india_cuisine = dict()

for i in delhi_ncr['Cuisines']:
    for j in i.split(", "):
        if j in delhi_ncr_cuisine:
            delhi_ncr_cuisine[j] += 1
        else:
            delhi_ncr_cuisine[j] = 1

for i in rest_of_india['Cuisines']:
    for j in i.split(", "):
        if j in rest_of_india_cuisine:
            rest_of_india_cuisine[j] += 1
        else:
            rest_of_india_cuisine[j] = 1

delhi_ncr_cuisine = sorted(delhi_ncr_cuisine.items(), key = lambda item: item[1], reverse = True)
rest_of_india_cuisine = sorted(rest_of_india_cuisine.items(), key = lambda item: item[1], reverse = True)

print('Top 10 cuisines served by maximum number of restaurants in Delhi NCR:')
for i in delhi_ncr_cuisine[0:10]:
    print(i[0])
print('--------------------------------------------------------------------------')
print('Top 10 cuisines served by maximum number of restaurants in rest of India:')
for i in rest_of_india_cuisine[0:10]:
    print(i[0])
```

### 1.4 Write a short detailed analysis of how cuisine served is different from Delhi NCR to Rest of India. Plot the suitable graph to explain your inference.

### 2.1 Write a short detail analysis of how the rating is affected by restaurant due following features: Plot a suitable graph to explain your inference.

#### 2.1.1  Number of Votes given Restaurant

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import pandas as pd
pd.options.mode.chained_assignment = None

data = pd.read_csv("zomato.csv", encoding = "ISO-8859-1")
data = data[data['Country Code'] == 1]
user_rating = data['Aggregate rating']
votes = data['Votes']
dict = {}
for i,j in zip(user_rating, votes):
    if i not in dict:
        dict[i] = [j,1]
    else:
        dict[i][0] += j
        dict[i][1] += 1
        
for i in dict.keys():
    dict[i] = dict[i][0]/dict[i][1]
dict = sorted(dict.items(), key = lambda item: item[0])

dict = np.array(dict)
user_rating = dict[:,1]
votes = dict[:,0]

plt.plot(votes, user_rating)
plt.xlabel('Aggregate Rating')
plt.ylabel('Votes')
plt.show()
```

### 2.2 Find the weighted restaurant rating of each locality and find out the top 10 localities with more weighted restaurant rating?

#### 2.2.1 Weighted Restaurant Rating= (number of votes * rating) /  (number of votes) .

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import pandas as pd
pd.options.mode.chained_assignment = None

data = pd.read_csv("zomato.csv", encoding = "ISO-8859-1")
data = data[data['Country Code'] == 1]
locality = data['Locality']
votes = data['Votes']
rating = data['Aggregate rating']
dict = {}

for i,j,k in zip(locality, votes, rating):
    if i not in dict:
        dict[i] = [j*k, j]
    else:
        dict[i][0] += j*k
        dict[i][1] += j

for i in dict.keys():
    if dict[i][0] != 0:
        dict[i] = round(dict[i][0]/dict[i][1], 2)
    else:
        dict[i] = 0

dict = sorted(dict.items(), key = lambda item: item[1])
dict = np.array(dict[-10:])
locality = dict[:, 0]
rating = dict[:,1]

plt.plot(locality, rating)
plt.xlabel('Locality')
plt.ylabel('Weighted Rating')
plt.xticks(rotation = 90)
plt.show()
```

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import pandas as pd
pd.options.mode.chained_assignment = None

data = pd.read_csv("zomato.csv", encoding = "ISO-8859-1")
data = data[data['Country Code'] == 1]
restaurant = data['Restaurant Name']
dict = {}

for i in restaurant:
    if i not in dict:
        dict[i] = 1
    else:
        dict[i] += 1

dict = sorted(dict.items(), key = lambda item: item[1])
dict = np.array(dict[-15:])
restaurant = dict[:, 0]
count = dict[:,1]

plt.bar(restaurant, count)
plt.xlabel('Restaurant')
plt.ylabel('No. of Outlets')
plt.xticks(rotation = 90)
plt.show()
```
